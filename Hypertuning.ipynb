```python
import pandas as pd
import  numpy as np
```

```python
data=pd.read_csv("pima-data.csv")
data.head(4)
```

```python
data.isnull().sum()
```

```python
# data['thickness'].value_counts()
```

```python
data['thickness']=np.where(data['thickness']==0,data['thickness'].mean(),data['thickness'])
```

```python
data.head(3)
```

```python
# data['diabetes']=np.where(data['diabetes']=='True',1,0)
data['diabetes']=np.where(data['diabetes']=='True',1,data['diabetes'])
```

```python
data.head(5)
```

```python
x=data.drop(['diabetes'],axis=1)
y=data['diabetes']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
cls=RandomForestClassifier()
cls.fit(x_train,y_train)
pred=cls.predict(x_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
print(accuracy_score(y_test,pred))
print(classification_report(y_test,pred))
print(confusion_matrix(y_test,pred))
```

```python
params={'n_estimators':[int(i) for i in np.linspace(100,1000,10)],
    'criterion':['entropy','gini'],
    'max_depth':[int(i) for i in np.linspace(100,1000,10)],
    'min_samples_split':[2, 5, 10,14],
    'min_samples_leaf':[1, 2, 4,6,8],
    'max_features':['auto', 'sqrt','log2'],
    'random_state':[10,20,30],
    
}
```

```python
from sklearn.model_selection import RandomizedSearchCV,GridSearchCV
random=RandomizedSearchCV(estimator=cls,param_distributions=params,n_iter=100,cv=3,verbose=2,
                               random_state=100)
```

```python
random.fit(x_train,y_train)
```

```python
random.best_params_
```

```python
random_cv=RandomForestClassifier(random_state= 30,
 n_estimators= 900,
 min_samples_split= 10,
 min_samples_leaf= 2,
 max_features= 'sqrt',
 max_depth= 100,
 criterion= 'entropy')
```

```python
random_cv.fit(x_train,y_train)
pred_randcv=random_cv.predict(x_test)
print(accuracy_score(y_test,pred_randcv))
print(classification_report(y_test,pred_randcv))
print(confusion_matrix(y_test,pred_randcv))
```

# Grid Search CV

```python
params={'random_state': [30,20],
 'n_estimators':[900],
 'min_samples_split': [10,20,30],
 'min_samples_leaf': [2],
 'max_features': ['sqrt'],
 'max_depth': [100,200,300],
 'criterion': ['entropy']}
```

```python
cls1=RandomForestClassifier()
grid=GridSearchCV( estimator=cls1,param_grid=params,cv=10,verbose=2)
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_estimator_
```

```python
grid.best_params_
```

```python
grid_cls=RandomForestClassifier(criterion='entropy', max_depth=100, max_features='sqrt',
                       min_samples_leaf=2, min_samples_split=10,
                       n_estimators=900, random_state=20)
grid_cls.fit(x_train,y_train)
grid_pred=grid_cls.predict(x_test)
```

```python
print(accuracy_score(y_test,grid_pred))
print(classification_report(y_test,grid_pred))
print(confusion_matrix(y_test,grid_pred))
```
